# Data Loading

## Import Library

```python
import pandas as pd
import numpy as np
from scipy.stats import skew
```

## Load Dataset

```python
dataset = pd.read_csv('Seasons_Stats.csv')
dataset.head()
```

----
# Data Preprocessing

## Get The Value in 2017

```python
dataset_2017 = dataset.loc[dataset['Year'] == 2017]
dataset_2017
```

## Reset Index The Column

```python
dataset_2017 = dataset_2017.reset_index(drop=True)
dataset_2017.head()
```

## Check The Column's Names

```python
dataset_2017.columns
```

## Delete All-Missing Value Column & Unnamed Column

```python
del dataset_2017['Unnamed: 0']
dataset_2017.head()
```

## Check The Missing Value for All Columns

```python
dataset_2017.isna().sum()
```

## Delete Columns "blanl" and "blank2"

```python
dataset_2017 = dataset_2017.drop(['blanl', 'blank2'], axis=1)
```

## Recheck The Column That Contains All Value Null

```python
dataset_2017.isna().sum()
```

## Check Duplicated Player's Name

```python
duplicated_player = dataset_2017[dataset_2017.duplicated('Player')]
duplicated_player.head(15)
```

## Drop Duplicated Player's Name

```python
dataset_2017.drop_duplicates(subset ="Player", keep = 'last', inplace = True)
dataset_2017.head(15)
```

## Reset Index

```python
dataset_2017 = dataset_2017.reset_index(drop=True)
dataset_2017
```

## Recheck The Null Value

```python
dataset_2017.isna().sum()
```

## Change The Null Value for Columns 'TS%', '3PAr', 'FTr', 'TOV%', 'FG%', '2P%', 'eFG%'

```python
dataset_2017.update(dataset_2017[['TS%', '3PAr', 'FTr', 'TOV%', 'FG%', '2P%', 'eFG%']].fillna(0.0))
```

## Recheck The Null Value

```python
dataset_2017.isna().sum()
```

## Check The Skewness of The Column '3P%'

```python
skew(dataset_2017['3P%'].dropna())
```

## Fill The NaN of Column '3P%' with Its Median

```python
dataset_2017['3P%'].fillna(dataset_2017['3P%'].median(), inplace=True)
dataset_2017['3P%'].isna().sum()
```

## Recheck The Null Value

```python
dataset_2017.isna().sum()
```

## Check The Skewness of The Column 'FT%'

```python
skew(dataset_2017['FT%'].dropna())
```

## Fill The NaN of Column 'FT%' with Its Median

```python
dataset_2017['FT%'].fillna(dataset_2017['FT%'].median(), inplace=True)
dataset_2017['FT%'].isna().sum()
```

## Recheck The Null Value

```python
dataset_2017.isna().sum()
```

# Answer The Problem Questions

## 1. Who is the youngest and oldest player in the NBA in 2017 for each team (Tm)?

```python
answer_1 = dataset_2017[['Player', 'Age', 'Tm']]
answer_1
```

### Find The Youngest Player in Each Team

```python
answer_1[['Tm', 'Player', 'Age']].groupby('Tm').min().reset_index()
```

### Find The Oldest Player in Each Team

```python
answer_1[['Tm', 'Player', 'Age']].groupby('Tm').max().reset_index()
```

## 2. Which player has the most minutes played (MP) in each position (Pos)?

```python
answer_2 = dataset_2017[['Player', 'MP', 'Pos']]
answer_2
```

```python
answer_2[['Pos', 'Player', 'MP']].groupby('Pos').max().reset_index()
```

## 3. Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?

```python
answer_3 = dataset_2017[['Tm', 'TRB%', 'AST%', 'STL%', 'BLK%']]
answer_3
```

```python
avg_answer_3 = answer_3.groupby(['Tm'], as_index=False)['TRB%', 'AST%', 'STL%', 'BLK%'].mean()
avg_answer_3
```

### Find The Highest Average TRB%

```python
avg_answer_3[avg_answer_3['TRB%'] == avg_answer_3['TRB%'].max()][['Tm','TRB%']].reset_index()
```

### Find The Highest Average AST%

```python
avg_answer_3[avg_answer_3['AST%'] == avg_answer_3['AST%'].max()][['Tm','AST%']].reset_index()
```

### Find The Highest Average STL%

```python
avg_answer_3[avg_answer_3['STL%'] == avg_answer_3['STL%'].max()][['Tm','STL%']].reset_index()
```

### Find The Highest Average BLK%

```python
avg_answer_3[avg_answer_3['BLK%'] == avg_answer_3['BLK%'].max()][['Tm','BLK%']].reset_index()
```

## 4. Who is the best player in your opinion based on his record stats?

To decide the best player, I choose the overall rating score based on these selected data, they are **AST, BLK, DRB, ORB, TRB, STL, TOV, PF, and PTS**. It is because these data is the inherent skill of the player.

```python
answer_4 = dataset_2017[['Player', 'AST', 'BLK', 'DRB', 'ORB', 'TRB', 'STL', 'TOV', 'PF', 'PTS']]
answer_4
```

### Calculate The Average Overall Score

```python
answer_4['Avg_Overall'] = answer_4.mean(axis=1)
answer_4
```

### Define The Best Player

```python
The_Best_Player = answer_4[answer_4['Avg_Overall'] == answer_4['Avg_Overall'].max()]
The_Best_Player
```

## 5.	Which team has the best average stat record of their players?

```python
answer_5 = dataset_2017[['Player', 'Tm', 'AST', 'BLK', 'DRB', 'ORB', 'TRB', 'STL', 'TOV', 'PF', 'PTS']]
answer_5
```

```python
answer_5['Avg_Overall'] = answer_4.mean(axis=1)
answer_5
```
